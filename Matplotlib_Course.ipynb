```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x) + np.random.normal(0,0.2,100)
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
