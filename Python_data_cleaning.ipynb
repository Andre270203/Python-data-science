This is a data cleaning practice

```python
#import pandas
import pandas as pd
```

```python
#import csv
df = pd.read_csv("/content/cricketdata.csv", sep = ";")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
#rename multiple columns
df = df.rename(columns = {"Mat" : "Matches","NO":"Not_Outs", "HS": "Highest_inns_Score", "BF": "Balls_Faced", "SR" : "Batting_Strike_Rate"})
```

```python
df.head()
```

```python
#chech null-values
df.isnull().any()
```

```python
df[df["Balls_Faced"].isna() == 1]
```

```python
df["Balls_Faced"] = df["Balls_Faced"].fillna(0)
```

```python
df["Batting_Strike_Rate"] = df["Batting_Strike_Rate"].fillna(0)
```

```python
df[df["Player"] == "ED Weekes (WI)"]
```

```python
#Drop duplicates
```

```python
df.duplicated().sum()
```

```python
df[df["Player"].duplicated() == 1]
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
#Split up Span into Start and End Date
df["Span"].str.split(pat="-")
```

```python
df["Rookie_Year"] = df["Span"].str.split(pat="-").str[0]
```

```python
df["Final_Year"] = df["Span"].str.split(pat="-").str[1]
```

```python
df.head()
```

```python
#Drop span column
df = df.drop(columns = "Span")
```

```python
df.head()
```

```python
#Split up the country from the player
df["Player"].str.split(pat="(")
```

```python
df["Country"] = df["Player"].str.split(pat="(").str[1]
```

```python
df["Country"] = df["Country"].str.split(pat=")").str[0]
```

```python
df["Player"] = df["Player"].str.split(pat="(").str[0]
```

```python
df.head()
```

```python
#Change Data Types
df.dtypes
```

```python
df["Highest_inns_Score"] = df["Highest_inns_Score"].str.split(pat="*").str[0]
```

```python
df["Highest_inns_Score"] = df["Highest_inns_Score"].astype("int")
```

```python
df.dtypes
```

```python
#Fix rookie_year and Final_Year
df = df.astype({"Rookie_Year": "int", "Final_Year":"int"})
df.dtypes
```

```python
#pd.set_option("display.max_rows", None)
```

```python
df
```

```python
#Remove FS.Jackson
df = df.drop(57, axis=0)
```

```python
df.isnull().any()
```

```python
df
```

```python
df["Balls_Faced"] = df["Balls_Faced"].str.split("+").str[0]
```

```python
df["Balls_Faced"] = df["Balls_Faced"].fillna(0)
```

```python
df
```

```python
df["Balls_Faced"] = df["Balls_Faced"].astype("int")
```

```python
df["Batting_Strike_Rate"] = df["Batting_Strike_Rate"].astype("float")
```

```python
df.dtypes
```

```python
#Build out Career_length column

df["Career_Length"] = df["Final_Year"] - df["Rookie_Year"]
df
```

```python
#Q.1 -> Cricketers in this df what is the average career length?
df["Career_Length"].mean()
```

```python
#Q.2 AVG Batting_Strike_Rate for cricketers who played over 10 years

df[df["Career_Length"] > 10]["Batting_Strike_Rate"].mean()
```

```python
#Q.3 find number of cricketers who played before 1960

df[df["Rookie_Year"] < 1960]["Player"].count()
```

```python
#Q.4 MAx highest Inns Score by country
df.groupby("Country")["Highest_inns_Score"].max()
```

```python
#Q.5 hunreds, fifties, ducks(0) AVG by Country
df.groupby("Country")[["100","50","0"]].mean()
```
