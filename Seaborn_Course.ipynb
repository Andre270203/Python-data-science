```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {"Category" : ["A","B","C","D","E"],
        "Value 1" : [15,30,5,10,20],
        "Value 2" : [10,25,8,15,30],
        "Value 3" : [5,10,15,20,25]}

df = pd.DataFrame(data)
df
```

```python
sns.barplot(x="Category",y="Value 1", data = df)
plt.title("Bar Plot")
plt.xlabel("Category")
plt.ylabel("Value 1")
plt.show()
```

```python
sns.scatterplot(data = df, x = "Value 1", y = "Value 2", hue = "Category")
plt.title("Scatter plot")
plt.xlabel("Value 1")
plt.ylabel("Value 2")
plt.legend(title = "Category")
plt.show()
```

```python
sns.lineplot(data = df, x = "Category", y = "Value 1", marker = "o")
plt.title("Line Plot")
plt.xlabel("Category")
plt.ylabel("Value 1")
plt.show()
```

```python
sns.heatmap(df[["Value 1","Value 2", "Value 3"]].corr(), annot = True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue = "Category", diag_kind = "kde")
plt.suptitle("Pairplot", y = 1.02)
plt.show()
```

```python
sns.countplot( x = "Category", data = df )
plt.title("Count Plot")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
sns.kdeplot(df[["Value 1","Value 2","Value 3"]], fill = True)
plt.title("KDE Plot")
plt.show()
```

```python
sns.regplot(x = "Value 1", y = "Value 2", data = df, marker = "o")
plt.title("Regplot")
plt.xlabel("Value 1")
plt.ylabel("Value 2")
plt.show()
```
